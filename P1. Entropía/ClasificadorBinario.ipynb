```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0.01,0.999,0.01)

# y = entropy
y = -x*np.log2(x)-(1.0-x)*np.log2(1.0-x)
```

Gini (x) = 1 - (x² + (1 - x)²)
 = 1 - (x² + 1 + x² - 2x)
 = 1 - x² - 1 - x² + 2x
 = 2x * (1 - x)

```python
gini = 2*x*(1-x)
```

```python
#plt.plot(x,y,x,gini)
plt.plot(x,y)
plt.plot(x,gini)
plt.grid()
plt.show()
```
