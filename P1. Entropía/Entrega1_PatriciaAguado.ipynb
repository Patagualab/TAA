Hallar la ganancia en información usando la entropía para cada uno de los cuatro atributos de la última tabla.

```python
import pandas as pd
import numpy as np
```

```python
# Lectura de datos
data = pd.read_csv('tennis.csv')

# Lectura de los atributos del fichero de datos
atributes = np.asarray(data.columns)

data
```

```python
data = np.asarray(data)
results = np.array([])
```

```python
# Función para el cálculo de la entropía asociada a la fuente
def entropy(data):
    h = 0
    n = data.shape[0]
    frecuency = np.unique(data[:,-1], return_counts= True)
    
    for i in frecuency[1]:
        h += (i/n)*(np.log2(n/i))
        
    return h
```

```python
# Función para el cálculo de la ganancia para cada atributo de los datos
def gain(atribute, data, results):
    h = 0
    gain = 0
    h_atr_index = 0
    
    
    # Datos de la clase
    c = data[:,4]
    
    # Selección de los datos del atributo
    atribute = data[:,atribute]
    n = atribute.shape[0]
    
    # Diferentes valores del atributo y su frecuencia de aparición
    frecuency = np.unique(atribute, return_counts= True)
    
    # Cálculo de la entropía para cada valor del atributo seleccionado
    for i in frecuency[0]:
        h_atr_index = 0
        index = np.where(atribute==i)[0]
        
        # Valores de clase para cada valor diferente del atributo
        class_index = c[index]
        
        # Frecuencia de cada valor de la clase para cada valor distinto del atributo
        f_atr_index = np.unique(class_index, return_counts= True)
        
        for j in f_atr_index[1]:
            h_atr_index += (j/len(class_index))*(np.log2(len(class_index)/j))
        
        
        h += (index.shape[0]/n)*h_atr_index
        
        
    # Ganancia del atributo pasado como parámetro
    gain = entropy(data)-h
    
    return gain   
```

```python
# Calculamos la ganancia de cada atributo
for i in range(len(atributes)-1):
    gain_atribute = gain(i,data, results)
    results = np.append(results,gain_atribute)
    
    print("\33[4mAtributo:\33[0m",atributes[i])
    print("\n  \033[1mGanancia:\033[0m %.3f" %gain_atribute,"\n")
    
max_gain = np.where(results == np.max(results))[0][0]

print("-------------------------------------------\n")
print("El atributo con mayor ganancia es:\033[1m",atributes[max_gain],"\033[0m")
```
