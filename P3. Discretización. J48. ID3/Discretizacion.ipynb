```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score as ac
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('ThoraricSurgery.csv')
data
```

```python
def discretize(data, column, interval):
    data_column = data[:,column]
    
    for d in range(len(data_column)):
        value = data_column[d]

        for i in range(len(interval[0])-1,-1,-1):
            
            if (value <= interval[0][i]):
                data_column[d] = interval[1][i]
                
            elif(interval[0][i] == interval[0][-1]):
                data_column[d] = interval[1][-1]
                
    data[:,column] = data_column
    return data
```

```python
data1 = np.asarray(data)
columns = [15,1,2]

intervals = [[[62,65,70],['62 o menos','De 63 a 65','De 66 a 70','Más de 70']],
            [[2.66,2.88],['2.66 o menos','De 2.67 a 2.88','Más de 2.88']],
            [[2.04],['2.04 o menos','Más de 2.04']]]
```

```python
for i in range(len(columns)):
    data = discretize(data1, columns[i], intervals[i])
print(data)
    
```
