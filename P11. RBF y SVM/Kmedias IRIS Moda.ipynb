```python
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.cluster import KMeans
from scipy.stats import mode
from sklearn.metrics import accuracy_score
```

```python
iris = datasets.load_iris()
y = iris.target
X = StandardScaler().fit_transform(iris.data)
print(np.unique(y))
```

```python
K = 10
cv_strat = StratifiedKFold(n_splits=K)
```

```python
max_clusters = 20
aciertos = np.zeros(max_clusters-2, dtype='float')

for k in range(3, max_clusters+1):
    print("\n\n---------------------- clusters ",k,"----------------------\n\n")
    kmedias = KMeans(n_clusters=k, random_state = 0, n_jobs=-1)
    
    for train_index, test_index in cv_strat.split(X,y):
        X_train = X[train_index]
        y_train = y[train_index]
        X_test = X[test_index]
        y_test = y[test_index]
        
        kmedias.fit(X_train)
        moda = -1 * np.ones(k, dtype='int')
        
        for i in range(k):
            print("cluster_en_carpeta",i,k)
            # Saca la moda para cada cluster por las clases del dataset
            moda[i] = mode(y_train[np.argwhere(kmedias.labels_ == i)])[0].astype(int)
        print(moda)
        a = kmedias.predict(X_test)
        print(a)
        y_test_predict = moda[a]
        print(y_test_predict)
        aciertos[k-3] += accuracy_score(y_test,y_test_predict)
    
    # Media de aciertos para cada cluster (media de los aciertos de cada carpeta)
    aciertos[k-3] /= K
    print(k,"-->", aciertos[k-3])
print(aciertos)
```
