```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score as ac
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('funcion.csv')
data
```

```python
data_feature_names = np.asarray(data.columns)
data_feature_names = np.delete(data_feature_names,-1)

data = np.asarray(data)
print(data)
data_class_names = np.unique(data[:,-1])
data_class_names = data_class_names.astype(str)
```

```python
X_train = data[:,:5]
y_train = data[:,-1]
```

```python
clf = DecisionTreeClassifier(random_state=0, criterion = 'entropy')
clf = clf.fit(X_train, y_train)

predicted = clf.predict(X_train)
ac = ac(predicted,y_train)

print("\n\33[4mTasa de acierto:\33[0m ",ac)
```

```python
error_rate = (1-ac)*100
print("\n\33[4mPorcentaje de ejemplos mal clasificados:\33[0m ",error_rate,"%")
```

```python
cm = confusion_matrix(y_train, predicted)
print("\n\33[4mMatriz de confusión:\33[0m\n")
for i in cm:
    print("    ",i)
print("\n")
```

```python
export_graphviz(clf, out_file='treeID3.dot', feature_names = data_feature_names, class_names = data_class_names, filled = True)
!dot -Tpng treeID3.dot -o treeID3.png -Gdpi=600
```
